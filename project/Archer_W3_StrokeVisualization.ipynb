```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import display, Markdown

todays_date = '2026-03-22'
student_id = 'halarc1407'

display(Markdown(f'**Date:** {todays_date}  |  **Student ID:** {student_id}'))
```

```python
# Load dataset
df = pd.read_excel('Stroke_Prediction_Dataset.xlsx')
print(f'Dataset shape: {df.shape}')
print(f'\nDiagnosis distribution:\n{df["Diagnosis"].value_counts()}')
print(f'\nNull values:\n{df.isnull().sum()[df.isnull().sum()>0]}')
df.head()
```

```python
# Step 3: Histogram — Age Distribution of All Patients
fig, ax = plt.subplots(figsize=(10, 5.5))
ax.hist(df['Age'], bins=30, color='#5B9BD5', edgecolor='white', linewidth=0.6)
ax.set_xlabel('Age', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of Patients', fontsize=12, fontweight='bold')
ax.set_title(f'Age Distribution of Patients\nStroke Prediction Dataset  |  {student_id}  |  {todays_date}', fontsize=13, fontweight='bold')
ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
ax.axvspan(18, 40, alpha=0.08, color='green', label='Younger adults (18-40)')
ax.axvspan(60, 90, alpha=0.08, color='orange', label='Older adults (60-90)')
ax.legend(fontsize=9, loc='upper right')
plt.tight_layout(); plt.show()
```

```python
# Step 4: Bar Chart — Work Type Distribution Among Stroke Patients
stroke_df = df[df['Diagnosis'] == 'Stroke'].copy()
work_counts = stroke_df['Work Type'].value_counts().reset_index()
work_counts.columns = ['work_type', 'count']
colors_bar = ['#2E75B6', '#5B9BD5', '#70AD47', '#ED7D31']

fig, ax = plt.subplots(figsize=(10, 5.5))
bars = ax.bar(work_counts['work_type'], work_counts['count'], color=colors_bar[:len(work_counts)], edgecolor='white', width=0.55)
ax.set_xlabel('Work Type', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of Stroke Patients', fontsize=12, fontweight='bold')
ax.set_title(f'Distribution of Work Types Among Stroke Patients\nStroke Prediction Dataset  |  {student_id}  |  {todays_date}', fontsize=13, fontweight='bold')
ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
for bar, val in zip(bars, work_counts['count']):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+8, str(val), ha='center', fontsize=11, fontweight='bold')
ax.set_ylim(0, work_counts['count'].max()*1.15)
plt.tight_layout(); plt.show()
```

```python
# Step 5: Scatter Plot — Age vs BMI for Stroke Patients
fig, ax = plt.subplots(figsize=(10, 5.5))
ax.scatter(stroke_df['Age'], stroke_df['Body Mass Index (BMI)'], alpha=0.35, c='#C00000', edgecolors='none', s=25, label='Stroke Patient')
z = np.polyfit(stroke_df['Age'], stroke_df['Body Mass Index (BMI)'], 1)
x_line = np.linspace(stroke_df['Age'].min(), stroke_df['Age'].max(), 100)
ax.plot(x_line, np.poly1d(z)(x_line), '--', color='#2E75B6', linewidth=2.5, label='Trend Line')
ax.set_xlabel('Age', fontsize=12, fontweight='bold')
ax.set_ylabel('Body Mass Index (BMI)', fontsize=12, fontweight='bold')
ax.set_title(f'Age vs. BMI Among Stroke Patients\nStroke Prediction Dataset  |  {student_id}  |  {todays_date}', fontsize=13, fontweight='bold')
ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
ax.legend(fontsize=10)
corr = stroke_df['Age'].corr(stroke_df['Body Mass Index (BMI)'])
ax.text(0.02, 0.95, f'Pearson r = {corr:.3f}\nn = {len(stroke_df):,} patients', transform=ax.transAxes, fontsize=10, va='top',
        bbox=dict(boxstyle='round,pad=0.4', facecolor='#EAF2FB', edgecolor='#2E75B6', alpha=0.9))
plt.tight_layout(); plt.show()
```

```python
# Step 6: Heatmap — Correlations Between Age, BMI, and Average Glucose Level
corr_df = df[['Age', 'Body Mass Index (BMI)', 'Average Glucose Level']].copy()
corr_df.columns = ['Age', 'BMI', 'Avg Glucose Level']
corr_matrix = corr_df.corr()

print('Correlation Matrix:'); display(corr_matrix.round(3))

fig, ax = plt.subplots(figsize=(7, 5.5))
sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='coolwarm', center=0, vmin=-1, vmax=1,
            linewidths=2, linecolor='white', annot_kws={'size': 14, 'weight': 'bold'},
            ax=ax, square=True, cbar_kws={'shrink': 0.8})
ax.set_title(f'Correlation Heatmap: Age, BMI, Avg Glucose Level\n{student_id}  |  {todays_date}', fontsize=12, fontweight='bold', pad=12)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha='right', fontsize=10)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=10)
plt.tight_layout(); plt.show()
```
